```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta = 1 / 1000
obs_all = []
```

```python
for _ in range(5):
    l1 = np.random.uniform(5, 10)
    l2 = l1 * np.exp(0.5)
    p1 = l1 * delta
    p2 = l2 * delta
    obs = np.hstack([np.random.binomial(450, p1, size=(100,)), np.random.binomial(450, p2, size=(300,))])
    obs_all.append(obs)
```

```python
for _ in range(5):
    l1 = np.random.uniform(5, 10)
    l2 = l1 * np.exp(-1)
    p1 = l1 * delta
    p2 = l2 * delta
    obs = np.hstack([np.random.binomial(450, p1, size=(100,)), np.random.binomial(450, p2, size=(300,))])
    obs_all.append(obs)
```
